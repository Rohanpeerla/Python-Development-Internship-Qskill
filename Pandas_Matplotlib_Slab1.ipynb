```python
#Load the File

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Hours_Studied": [2, 4, 6, 8, 10],
    "Marks": [40, 55, 75, 85, 95]
}

df = pd.DataFrame(data)
df
```

```python
average_marks = df["Marks"].mean()
print("Average Marks:", average_marks)
```

```python
#Bar Chart
plt.bar(df["Name"], df["Marks"])
plt.xlabel("Name")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.show()
```

```python
#Scatter Plot
plt.scatter(df["Hours_Studied"], df["Marks"])
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Hours Studied vs Marks")
plt.show()
```

```python
#Heatmap
correlation = df[["Hours_Studied", "Marks"]].corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

```

Insights and Observations
. The Average Marks show overall student performance.
. Bar Chart Helps compare marks between students.
. Scatter plot shows that marks increase as study hours increase.
. Heatmap shows strong positive correlation between hours studied and marks.
