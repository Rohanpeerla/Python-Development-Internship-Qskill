````python
import google.generativeai as genai
import requests
from google.colab import userdata

GEMINI_API_KEY = userdata.get("GEMINI_API_KEY")
SERPAPI_KEY = userdata.get("SERPAPI_KEY")

genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel("gemini-flash-latest")

conversation_history = []

def real_time_search(query):
    url = "https://serpapi.com/search"
    params = {
        "engine": "google",
        "q": query,
        "api_key": SERPAPI_KEY
    }
    data = requests.get(url, params=params).json()

    if "answer_box" in data and "answer" in data["answer_box"]:
        return data["answer_box"]["answer"]
    elif "organic_results" in data:
        return data["organic_results"][0]["snippet"]
    else:
        return "No real-time data found."

def gemini_chat(user_input):
    conversation_history.append(f"User: {user_input}")
    prompt = "\n".join(conversation_history)

    response = model.generate_content(prompt)
    reply = response.text

    conversation_history.append(f"Assistant: {reply}")
    return reply

print("ðŸ¤– Gemini Assistant Started (type 'exit' to stop)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Assistant: Goodbye ðŸ‘‹")
        break

    if "price" in user_input.lower() or "weather" in user_input.lower():
        ans = real_time_search(user_input)
        print("Assistant (Real-Time):", ans)
        conversation_history.append(f"Assistant: {ans}")
    else:
        print("Assistant:", gemini_chat(user_input))
````
